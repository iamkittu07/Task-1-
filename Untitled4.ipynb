```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def region_of_interest(image):
    height = image.shape[0]
    polygons = np.array([
        [(100, height), (image.shape[1]-100, height), (image.shape[1]//2, height//2)]
    ])
    mask = np.zeros_like(image)
    cv2.fillPoly(mask, polygons, 255)
    masked_image = cv2.bitwise_and(image, mask)
    return masked_image
def draw_lines(image, lines):
    line_image = np.zeros_like(image)
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(line_image, (x1, y1), (x2, y2), (0, 255, 0), 5)
    return line_image
image = cv2.imread('image.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5, 5), 0)
edges = cv2.Canny(blur, 50, 150)

# Focus on region of interest
roi = region_of_interest(edges)

# Detect lines using Hough Transform
lines = cv2.HoughLinesP(roi, 1, np.pi / 180, threshold=50, minLineLength=50, maxLineGap=150)

# Draw lines on the original image
line_image = draw_lines(image, lines)
combo = cv2.addWeighted(image, 0.8, line_image, 1, 1)

# Show the result
combo = cv2.cvtColor(combo, cv2.COLOR_BGR2RGB)
plt.imshow(combo)
plt.axis('off')
plt.title("Detected Lane Lines")
plt.show()
```
